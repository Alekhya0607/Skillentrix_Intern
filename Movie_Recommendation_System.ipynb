```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
## load dataset
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv.zip', compression='zip')

movies.head()
```

```python
## selecting useful columns

movies = movies[['movieId', 'title', 'genres']]
movies.head()
```

```python
## Filling missing values
movies.fillna('', inplace=True)
```

```python
## Converting text data
tfidf = TfidfVectorizer(stop_words='english')

tfidf_matrix = tfidf.fit_transform(movies['genres'])
```

```python
## calculate similarity
cosine_sim = cosine_similarity(tfidf_matrix)
```

```python
## Creating Recommendation function
def recommend_movies(movie_name):

    idx = movies[movies['title'] == movie_name].index[0]

    similarity_scores = list(enumerate(cosine_sim[idx]))

    similarity_scores = sorted(similarity_scores,
                               key=lambda x: x[1],
                               reverse=True)

    similarity_scores = similarity_scores[1:11] # Changed from 1:6 to 1:11 for top 10 recommendations

    for movie in similarity_scores:
        print(movies.iloc[movie[0]].title)
```

```python
## Testing the system
recommend_movies('Jumanji (1995)')
```

```python
## User input
movie_name = input("Enter a movie name: ")

# Check if 'movies' DataFrame exists and if the movie is found in it.
# This prevents the IndexError if the movie name is not in the dataset.
if 'movies' in globals() and movie_name in movies['title'].values:
    recommend_movies(movie_name)
else:
    print(f"Movie '{movie_name}' not found in the dataset or 'movies' DataFrame is not loaded.")
    print("Please ensure you enter an exact movie title (e.g., 'Toy Story (1995)').")
    print("Also, make sure all previous cells, especially the one defining the 'recommend_movies' function (cell 'qkOC396dK0qw'), have been run.")
```
